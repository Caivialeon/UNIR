```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```python
# load the image
img = load_img('./drive/MyDrive/SCA_test/img1.jpg')

# convert image to numpy array
data = img_to_array(img)

# reshape (1,180,270,3)
samples = data.reshape((1,)+data.shape)
```

```python
'''
  Create an instance of the ImageDataGenerator
  horizontal_flip and vertical_flip - boolean
  width_shift_range and height_shift_range - real number [0,1] or range
  brightness_range - range between 0 and 1
  rotation_range - value between 0 and 180
  zoom_range - 
'''
datagen = ImageDataGenerator(
    #horizontal_flip=True,
    #vertical_flip=True,
    #rotation_range=90,
    #brightness_range=[0.1, 0.8]
    #zoom_range=[0.1,1.0],
    width_shift_range=0.5,
    #height_shift_range=0.1
    #shear_range=0.7,
    fill_mode='nearest'
)
'''
  For an image dataset loaded in memory an iterator must be created
'''
it = datagen.flow(samples, batch_size=1)

for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
# show the figure
pyplot.show()
```
