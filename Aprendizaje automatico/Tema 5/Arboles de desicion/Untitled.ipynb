Regresion con Arboles de desicion

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor , export_graphviz
from sklearn.model_selection import GridSearchCV
#Visualizacion
import matplotlib.pyplot as plt
import seaborn as sns
import graphviz
%matplotlib inline
#metricas
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

Cargamos los datos

```python
#cargamos el conjunto de datos de viviendas de Boston desde Sklearn.datasets
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
datos_x = pd.DataFrame(boston.data, columns = boston.feature_names)
datos_y = pd.DataFrame(boston.target)
datos_x.describe()
```

Determinar el conjunto de modelizacion y el de validacion
*Se Segmenta los datos en dos conjuntos, uno de entrenamiento con 80%  y de prueba 20%  del total de datos respectivamente

```python
#dividir conjuntos train/test
x_train , x_test ,y_train ,y_test = train_test_split(datos_x , datos_y , test_size = 0.20 , random_state = 0)
```

Definir un arbol de regresion empleando como variable respuesta "MEDV" y como predictores todas las otras variables

Parametrizacion del algoritmo de regresion

```python
#entrenamos el modelo
modelo = DecisionTreeRegressor(max_depth = 6 , splitter = "best", random_state=123)
modelo.fit (x_train,y_train)
```

Estructura del arbol

```python
from sklearn.tree import plot_tree
fig,ax = plt.subplots(figsize = (12,5))

print(f"profundidad del arbol: {modelo.get_depth()}")
print(f"Numero de nodos terminales : {modelo.get_n_leaves()}" )

plot = plot_tree(
                decision_tree=modelo,
                feature_names=x_test.columns,
                class_names="MEV",
                filled = True,
                impurity = False,
                fontsize = 11,
                precision = 2,
                ax = ax
)
```

Prediccion del modelo

```python
prediccion = modelo.predict(x_test)
```

Metricas

*Evaluacion del modelo

```python
print("MAE" , mean_absolute_error(y_test , prediccion))
print("MSE" , mean_squared_error(y_test , prediccion))
print("R2" , r2_score(y_test,prediccion))
```

Podado del arbol

ccp_alpha determina el grado de penalizacion por complejidad

```python
#valores de ccp_alpha evaluados
param_grid = {"ccp_alpha" :np.linspace(0,50,20)}
#busqueda por validacion cruzada
grid = GridSearchCV(
                #el arbol se crece al maximo posible para luego aplicar el pruning
                estimator = DecisionTreeRegressor(
                                                max_depth = None,
                                                min_samples_split=20,
                                                min_samples_leaf=1,
                                                random_state=123
                                                ),
                param_grid = param_grid,
                cv = 10,
                refit = True,
                return_train_score = True
                )

grid.fit (x_train , y_train)

fig , ax = plt.subplots (figsize = (6,3.84))
scores = pd.DataFrame(grid.cv_results_)
scores.plot(x="param_ccp_alpha" , y = "mean_train_score" , yerr="std_train_score" , ax=ax)
scores.plot(x = "param_ccp_alpha" , y = "mean_test_score" , yerr = "std_test_score" , ax=ax)
ax.set_title("Error de validacion cruzada vs hiperparametro ccp_alpha" , color = "red");
```

Estructura del arbol final

```python
modelo_final = grid.best_estimator_
print(f"Profundidad del arbol : {modelo_final.get_depth()}")
print(f"Numero de nodos terminales :  {modelo_final.get_n_leaves()}" )

fig,ax = plt.subplots(figsize = (7,5))
plot = plot_tree(
                decision_tree = modelo_final,
                feature_names = x_test.columns,
                class_names = "MEDV",
                filled = True,
                impurity = False,
                ax = ax
)
```

Prediccion del modelo

```python
prediccion_final = modelo_final.predict(x_test)
```

Metricas
*Evaluacion del modelo

```python
print("MAE" , mean_absolute_error(y_test , prediccion_final ))
print("MSE" , mean_squared_error(y_test , prediccion_final))
print("R2" , r2_score(y_test,prediccion_final))
```
